## Background & Prior Work




The Central Valley of California is known for being an agricultural heartland as well as being  prone to significant ozone pollution. According to the American Lung Association's State of the Air Report, two cities in the Central Valley are ranked amongst the top five most polluted cities in the country.<sup>1</sup> The cause of pollution in the Central Valley can be attributed to its topography that traps air inside the region, in conjunction with its growing population and industry.<sup>2</sup> Despite there having been statewide regulations placed to decrease agricultural emissions in 2004, ozone pollution still remains significant.<sup>3</sup> 

The central valley has repeatedly made efforts to combat the rising levels of pollution, but it has been met with less than desirable results. One example was the Spare The Air program which was more focused on the actions of the individual encouraging people to drive less among other things during high ozone days, but this didn’t really tackle the source of the problem. Also going back to 2007 a new plan known as the Ozone Plan came into existence and by its guidelines places like the central valley would take till 2024 under the Extreme Non-Compliance standard before they reached full ozone compliance. This led to the creation of the ISSRC in the central valley, so it could be decided if there were any alternatives to the 2024 plan. This inevitably led to a lot of back and forth between science and politics as different parties believed in different solutions to the ozone problem in the Central Valley. <sup>4</sup> Thankfully, the Central Valley hasn’t given up so easily with further efforts being made to combat the problem with a new version of The Ozone Plan being introduced in 2022.<sup>5</sup>

Air pollution can be a cause for concern with regards to respiratory health. Increased pollution rates have been found to cause an increase in respiratory symptoms, a decrease in pulmonary function onset of asthma, and increased mortality and hospitalizations. <sup>6</sup> In a study across Europe, it was also found that living near traffic-heavy roads could potentially increase a person’s risk of lung cancer.<sup>7</sup> We hope to be able to visualize these effects of air pollution on respiratory health in the counties of the California’s Central Valley.

## Data Cleaning: Primary Care Shortage Areas

```python
#load primary care shortage areas csv
pcsa = pd.read_csv('https://data.chhs.ca.gov/dataset/061494a3-e8c7-4615-a22f-b2851d44eb09/resource/0ba7c904-2302-400a-ba27-b8e8e5c1ab4a/download/pcsa.csv')
pcsa.head()
```

```python
#filter out the counties to only get counties in the central valley
pcsa_sub = pcsa.loc[(pcsa['MSSA_COUNTY'] == 'Butte') | (pcsa['MSSA_COUNTY'] == 'Glenn') | 
(pcsa['MSSA_COUNTY'] == 'Fresno') | (pcsa['MSSA_COUNTY'] == 'Kern') | 
(pcsa['MSSA_COUNTY'] == 'Kings') | (pcsa['MSSA_COUNTY'] == 'Madera') | 
(pcsa['MSSA_COUNTY'] == 'Merced') | (pcsa['MSSA_COUNTY'] == 'Placer') | 
(pcsa['MSSA_COUNTY'] == 'San Joaquin') | (pcsa['MSSA_COUNTY'] == 'Sacramento') | 
(pcsa['MSSA_COUNTY'] == 'Shasta') | (pcsa['MSSA_COUNTY'] == 'Solano') | 
(pcsa['MSSA_COUNTY'] == 'Stanislaus') | (pcsa['MSSA_COUNTY'] == 'Sutter') | 
(pcsa['MSSA_COUNTY'] == 'Tehama') | (pcsa['MSSA_COUNTY'] == 'Tulare')
]
# only keep columns for county, score provider ratio, poverty score, total score, PCSA & Population
pcsa_sub = pcsa_sub[['MSSA_COUNTY', 'Score_Provider_Ratio', 'Score_Poverty', 'Score_Total', 'PCSA', 'Total_Population']]

# rename county column
pcsa_sub.columns = ['County','Score_Provider_Ratio', 'Score_Poverty','Score_Total', 'PCSA', 'Population']

# replace Yes and No PCSA values with binary values
def replace_with_number(value):
   
    value = value.replace("No", 0)
    value = value.replace("Yes", 1)
    
    return value
pcsa_sub = pcsa_sub.apply(replace_with_number)

# get average score total and number of PCSA cities/areas within the county 
list_counties = pcsa_sub.County.unique()
pcsa_county = pd.DataFrame(
                   columns=['County', 'Score_Provider_Ratio', 'Score_Poverty', 'Score_Total', 'PCSAs_Total','Population'])
for county in list_counties:
    # add total provider ratio for each county
    provider_rate_total = pcsa_sub.loc[(pcsa_sub['County'] == county)]['Score_Provider_Ratio'].sum()
    # add total poverty score for each county
    poverty_score_total = pcsa_sub.loc[(pcsa_sub['County'] == county)]['Score_Poverty'].sum()
    # add total scores for each county
    ind_score_total = pcsa_sub.loc[(pcsa_sub['County'] == county)]['Score_Total'].sum()
    # add total scores for each county
    ind_score_total = pcsa_sub.loc[(pcsa_sub['County'] == county)]['Score_Total'].sum()
    # add number of PCSA cites/areas
    pcsa_total = pcsa_sub.loc[(pcsa_sub['County'] == county)]['PCSA'].sum()
    # find total population for each county
    pop_total = pcsa_sub.loc[(pcsa_sub['County'] == county)]['Population'].sum()

    #add all above values to the dataframe
    pcsa_county = pcsa_county.append({'County': county,'Score_Provider_Ratio': provider_rate_total, 
                                        'Score_Poverty': poverty_score_total, 'Score_Total': ind_score_total, 
                                        'PCSAs_Total': pcsa_total, 'Population': pop_total}, ignore_index=True)

# change numerical object values to float
pcsa_county["PCSAs_Total"] = pcsa_county["PCSAs_Total"].astype(float)
pcsa_county["Score_Provider_Ratio"] = pcsa_county["Score_Provider_Ratio"].astype(float)
pcsa_county["Score_Poverty"] = pcsa_county["Score_Poverty"].astype(float)
pcsa_county["Score_Total"] = pcsa_county["Score_Total"].astype(float)
pcsa_county["Population"] = pcsa_county["Population"].astype(float)
pcsa_county.head()
```

We then plotted the average AQI across 2014-2018 for each county by the county's lung cancer rate per 100,000 residents.

```python
# get average AQI across years for each county & append values to AQI data frame
counties = ['Butte', 'Fresno', 'Glenn', 'Kern', 'Kings', 'Madera', 
            'Merced', 'Placer', 'San Joaquin', 'Sacramento', 'Shasta', 'Solano', 'Stanislaus', 'Sutter', 'Tehama', 'Tulare']
avgAQI = []

for county in counties:
        all_county_aqi = aqiTotal[(aqiTotal['County'] == county)]
        countyAvg = all_county_aqi[["AQI"]].mean()
        avgAQI.append(countyAvg[0])

#append AQI avgs to cancer df
cancer_sub.reset_index(drop = True)
cancer_sub['averageAQI'] = avgAQI
sns.set(style="white", font_scale = 2)
cancervsaqiavg = sns.regplot(x = "averageAQI", y = "Age_adjusted_rate", data = cancer_sub)
cancervsaqiavg.figure.set_size_inches(8,8)
cancervsaqiavg.set_title('AQI Average vs. Lung Cancer Rate')
cancervsaqiavg.set_ylabel('Lung Cancer rate per 100,000')
```

There looks to be a negative correlation given by the steep downward line, though the points are widely scattered.

We then ran a linear regression analysis and t-test via OLS. 

```python
outcome, predictors = patsy.dmatrices('Age_adjusted_rate ~ averageAQI', cancer_sub)
mod = sm.OLS(outcome, predictors)
canceraqiplot = mod.fit()
print(canceraqiplot.summary())
```

The regression shows an r-squared of 0.514, which is not necessarily strong. Where we see an important statistic is the p-value of 0.002, which demonstrates that __there is a significant effect of AQI on the rate of lung cancer__ as it is much less than 0.05.

We then plotted the average AQI across 2014-2018 for each county by the county's rate of death by resp. disease (a ratio of total deaths by resp. diseease to population). 

```python
# get average AQI across years for each county & append values to AQI data frame
deaths_and_AQI = pd.DataFrame()
avgDEATHS = []

for county in counties:
        all_county_deaths = deaths_by_respiratory_per_county_pop[(deaths_by_respiratory_per_county_pop['County'] == county)]
        countydeathAvg = all_county_deaths[["Count"]].mean()
        avgDEATHS.append(countydeathAvg[0])

#append AQI avgs to death/AQI df
deaths_and_AQI['county'] = counties
deaths_and_AQI['averageAQI'] = avgAQI
deaths_and_AQI['averageDeaths'] = avgDEATHS
deaths_and_AQI['population'] = cancer_sub["Population"].tolist()
deaths_and_AQI['deathRate'] = deaths_and_AQI['averageDeaths'] / deaths_and_AQI['population']

# plot line
sns.set(font_scale=2, style="white")
deathavgvsaqiavg = sns.regplot(x = "averageAQI", y = "deathRate", data = deaths_and_AQI)
deathavgvsaqiavg.set_title('AQI Average vs. Death Rate')
deathavgvsaqiavg.set_ylabel('Average Death Rate')
deathavgvsaqiavg.figure.set_size_inches(8,8)
```

There looks to be a somewhat negative correlation given by the downward line, though the points are widely scattered and the line is not very steep.

We then ran a linear regression analysis and t-test via OLS.

```python
outcome, predictors = patsy.dmatrices('deathRate ~ averageAQI', deaths_and_AQI)
mod = sm.OLS(outcome, predictors)
deathsaqiplot = mod.fit()
print(deathsaqiplot.summary())
```

### Predictive Model on Lung Cancer Rates

This part of our predictive analysis focused on trying to use the average AQI for each respective county along with their poverty score to try and predict what category each county would fall into for lung cancer rates

```python
#provided function from D8 on trainging SVMs
def train_SVM(X, y, kernel='linear'):
    clf = SVC(kernel=kernel)
    clf.fit(X, y)
    
    return clf
```

```python
#We set up our intial prediction dataframe with the average AQI, poverty score, and the adjusted lung cancer rates
pred_df = pd.DataFrame()
pred_df["County"] = pcsa_and_death["County"]
pred_df["Poverty_Score"] = pcsa_and_death["Poverty_Score"]
tmp = cancer_sub["Age_adjusted_rate"].tolist()
pred_df["Age_adjusted_rate"] = tmp
tmp = cancer_sub["averageAQI"].tolist()
pred_df["averageAQI"] = tmp
pred_df
```

```python
#we set up the training and testing data sets 
num_training = int(len(pred_df) * 0.8)
num_testing = int(len(pred_df) - num_training)

#our predicted variable needs to be categorical so we're going to break them into categories based off of the number of lung cancer reports
categLung = []
ind = 0
while(ind < 16):
    tmp = pred_df.iat[ind, 2]
    #print(pred_df['Age_adjusted_rate'][ind])
    if(tmp >= 36.4 and tmp < 41.8):
        categLung.append("least prominent")
    elif(tmp >= 41.8 and tmp < 47.95):
        categLung.append("somewhat prominent")
    elif(tmp >= 47.95 and tmp < 54.2):
        categLung.append("more prominent")
    else:
        categLung.append("most prominent")
    ind += 1

pred_df["Categorical_Lung _Cancer_Rate"] = categLung

#we are going to try and predict cancer rates from aqi rates and poverty scores 
pred_X = pred_df[["averageAQI", 'Poverty_Score']]
pred_Y = np.array(pred_df['Categorical_Lung _Cancer_Rate'])

#We split our data into the apropiate testing and training sets
pred_train_X = pred_X[:num_training]
pred_train_Y = pred_Y[:num_training]
pred_test_X = pred_X[num_training:]
pred_test_Y = pred_Y[num_training:]

#Initiate our SVM
pred_clf = train_SVM(pred_train_X, pred_train_Y)

#we make our predictions on the provided training data
predicted_train_Y = pred_clf.predict(pred_train_X)
predicted_test_Y = pred_clf.predict(pred_test_X)

#Produce report on the training data
class_report_train = None
class_report_train = classification_report(pred_train_Y, predicted_train_Y)
print(class_report_train)

```
